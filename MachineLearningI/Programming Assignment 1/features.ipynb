```
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
from sklearn import preprocessing
import scipy
from scipy.stats.stats import pearsonr
 

 
```

```
dataframe =  pd.read_csv('/content/drive/My Drive/Colab Notebooks/dataset.csv')
features = dataframe.iloc[:,0:-1]
y = dataframe.iloc[:,-1]
#print(features)
#print(y)
```

```
nfeatures = preprocessing.normalize(features)
np.savetxt("/content/drive/My Drive/Colab Notebooks/NormalizeData.csv", nfeatures, delimiter=",")
```

```
#Calculating the coefficient value for each featureset
p_values = []
for i in range(12):
  x = dataframe.iloc[:,i]
  #print(x)
  #print(y)
  coeff, p = pearsonr(x, y)
  print(coeff)
  p_values.append(coeff)
```

```
#top 5 features are manually loaded into a new CSV top5.csv


dataframe_min =  pd.read_csv('/content/drive/My Drive/Colab Notebooks/top5.csv')
data_min = dataframe_min.iloc[:,0:-1]
#y = dataframe_min.iloc[:,-1]
title = data_min.columns.values.tolist()
print(title)

```

```
#Correlation graphs for the top 5 input features

for i in range(5):
  xlabel = title[i]
  x = dataframe_min.iloc[:,i]
  %matplotlib inline
  matplotlib.style.use('ggplot')  

  plt.scatter(x,y,alpha=0.5)
  
  plt.title('Correlation graph of ' + xlabel + ' with respect to Area')
  plt.xlabel(xlabel)
  plt.ylabel('Area')
  plt.show()
```
