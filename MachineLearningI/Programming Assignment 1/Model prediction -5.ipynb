```
import numpy as np
import pandas as pd
import math
#import matplotlib
#import matplotlib.pyplot as plt
from numpy.linalg import inv
#from sklearn import preprocessing
from google.colab import drive
drive.mount('/content/drive/')
```

```
def calculate_Beta(x,y):
 
  x_trans = pd.DataFrame(np.transpose(x))
  
  intermediate = pd.DataFrame(np.dot(x_trans,x))
  
  inverse = inv(np.matrix(intermediate))
  
  dotproduct = pd.DataFrame(np.dot(x_trans, y))
  
  beta = pd.DataFrame(np.dot(inverse,dotproduct))
  
  return beta
```

```
def calculate_ycap(x, beta):
  
  #beta = calculate_Beta(x)

  ycap = pd.DataFrame(np.dot(x,beta))
 
  return ycap
```

```
def calculate_MAE(x, y, ycap):
  
  sum = 0
  
  for i in range(len(x)):
    
    y_i = y.iloc[i]
    
    y_i_cap = ycap.iloc[i]
    
    difference = abs(y_i_cap - y_i)
    
    sum += difference
    
  return (sum/len(x))
```

```
def calculate_RMSE(x, y, ycap):
  sum = 0
  for i in range(len(x)):
    y_i = y.iloc[i]
    y_i_cap = ycap.iloc[i]
    squared = (y_i_cap - y_i) **2
    sum += squared
  avg = sum/len(x)
  return (math.sqrt(avg))
```

```
MAE = []
RMSE = []
data_source = '/content/drive/My Drive/Colab Notebooks/FCV/Top 5/Model 1/'

for val in range(1,11):
  training_data = pd.read_csv(data_source + 'train' +str(val) + '.csv')
  training_y = training_data.iloc[:,-1]
  training_x = training_data.iloc[:,0:-1]
  
  testing_data = pd.read_csv(data_source + 'test' +str(val) + '.csv')
  testing_y = testing_data.iloc[:,-1]
  testing_x = testing_data.iloc[:,0:-1]
  
  beta = calculate_Beta(training_x, training_y)
  y_predict = calculate_ycap(training_x, beta)
  
  MAE_val = calculate_MAE(testing_x, testing_y, y_predict)
  
  MAE.append(MAE_val)
  
  RMSE_val = calculate_RMSE(testing_x, testing_y, y_predict)
  
  RMSE.append(RMSE_val)
  

MAE_mean = np.mean(MAE)
MAE_sd = np.std(MAE)

RMSE_mean = np.mean(RMSE)
RMSE_sd = np.std(RMSE)

print("For Model 1: \n")
print("Mean of MAE is %f" % MAE_mean)
print("SD of MAE is %f" % MAE_sd)
print("Mean of RMSE is %f" % RMSE_mean)
print("SD of RMSE is %f" % RMSE_sd)


```
